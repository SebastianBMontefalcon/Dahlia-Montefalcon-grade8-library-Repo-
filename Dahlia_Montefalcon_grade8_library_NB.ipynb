```python
import requests
import json

url = "https://raw.githubusercontent.com/SebastianBMontefalcon/Dahlia-Montefalcon-grade8-library-Repo-/refs/heads/main/books.json"
response = requests.get(url)
books_data_list = response.json() # Renamed 'books' to 'books_data_list' to avoid confusion with the single book dictionary 'books'

print("Books data loaded:")
print(books_data_list)

total_books_in_library = 0
for book in books_data_list['books']:
    total_books_in_library += book['details']['copies']

print(f"\nTotal books (all copies combined) in the library: {total_books_in_library}")

#make sure the url is the raw file from the github
```
